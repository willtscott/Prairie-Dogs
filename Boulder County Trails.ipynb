## Data Cleanup

```python
# Convert Categorical Features
# veg_type = pd.get_dummies(data['VEG_TYPE'],drop_first=True)
# data = pd.concat([data,veg_type],axis=1)
```

```python
# Drop trivial columns
data = colonies_old.drop(['OBJECTID', 'ASPECT', 'Town', 'CATEGORY_L','PROPERTY', 'INFLUENCE', 'LANDUSE','VEG_TYPE','STATUS', 'PDOG_CAT', 'REMARKS','RELOCATION', 'Encroaching','ShapeSTArea'], axis=1)
```

```python
def yesNo(str):
    return int(str.lower() == 'yes')
```

```python
data['AG_LEASE'] = data['AG_LEASE'].fillna('no')
data['AG_LEASE'] = data['AG_LEASE'].apply(yesNo)
```

```python
data['YOUNG_EMER'] = data['YOUNG_EMER'].fillna('no')
data['YOUNG_EMER'] = data['YOUNG_EMER'].apply(yesNo)
```

```python
def activeInactive(str):
    return int(str.lower() == 'active')
```

```python
data['ACTIVE'] = data['ACTIVE'].fillna('inactive')
data['ACTIVE'] = data['ACTIVE'].apply(activeInactive)
```

```python
data.dropna(inplace=True)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(data.corr())
```

# Machine Learning

## Linear Regression on Acreage Feature

```python
sns.lmplot('GPS_YEAR', 'Acreage', data=data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = data.drop('Acreage', axis=1)
y = data['Acreage']
```

```python
lin_r = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lin_r.fit(X_train, y_train)
```

```python
coeff_df = pd.DataFrame(lin_r.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

Interpreting the coefficients:

- Holding all other features fixed, a 1 unit increase in **HOLES** is associated with an **increase of .001396 Acres **.
- Holding all other features fixed, a 1 unit increase in **DENSITY** is associated with an **increase of .059495 Acres **.
- Holding all other features fixed, a 1 unit increase in **ShapeSTLength** is associated with an **increase of .005602 Acres **.

Do these make sense? Maybe, but it's hard to tell because so few of the features are linear.

```python
predictions = lin_r.predict(X_test)
```

### Evaluation

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

## Regression on Active Feature

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = data.drop('ACTIVE', axis=1)
y = data['ACTIVE']
```

```python
lm = LogisticRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lm.fit(X_train, y_train)
```

```python
predictions = lm.predict(X_test)
```

### Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
